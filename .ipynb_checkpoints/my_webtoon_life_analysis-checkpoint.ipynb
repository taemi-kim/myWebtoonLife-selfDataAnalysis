## Í∑∏Îü¨ÎÇò, Ïó¨Í∏∞ÏÑú ÏÉùÍ∞ÅÌï¥ÏïºÌï† Í≤ÉÏùÄ ÎÇ¥Í∞Ä Î≥¥Îäî ÏõπÌà∞ Ï§ë Ìï¥Îãπ Ïû•Î•¥Í∞Ä Ï∞®ÏßÄÌïòÎäî ÎπÑÏú®Ïù¥Îã§. ### 2-1c. ÌååÏÉùÎ≥ÄÏàòÎ•º Í∞ÄÏ§ëÏπò ÏÇºÏïÑ Îòê ÌïòÎÇòÏùò ÌååÏÉùÎ≥ÄÏàòÎ•º ÎßåÎì†Îã§Î©¥?

```python
#Ïû•Î•¥ ÎπàÎèÑÏàò / Ïû•Î•¥ Ï¥ùÌï© = Ïû•Î•¥ ÎπÑÏú®
#Ï≤òÏùåÏóê Í≥ÑÏÇ∞ÌñàÎçò .value_counts()Í∞íÏúºÎ°ú ÎπÑÏú®ÏùÑ Í≥ÑÏÇ∞ÌïúÎã§ 
df_genre_counts = pd.DataFrame(round(genre_counts/genre_counts.sum(),3))
df_genre_counts = df_genre_counts.rename(columns={"Ïû•Î•¥":"Ïû•Î•¥ ÎπÑÏú®"})

#df_genre_counts = df_genre_counts.reset_index()
df_genre_counts = df_genre_counts.reset_index().rename(columns={"index":"Ïû•Î•¥"})
df_genre_counts
```

```python
df_genre
```

```python
#Ïû•Î•¥Î°ú Ìï©Ïπú ÌõÑ Í≥ÑÏÇ∞ÌïòÏó¨ ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ± 
df_genre_final = pd.merge(df_genre_counts, df_genre, how='inner', on='Ïû•Î•¥')
df_genre_final["ÏÑ†Ìò∏ÎèÑ_weighted"] = df_genre_final["Ïû•Î•¥ ÎπÑÏú®"] * df_genre_final["ÏÑ†Ìò∏ÎèÑ"]
df_genre_final
```

```python
plt.figure(figsize=(16,8))
plt1 = sns.barplot(data=df_genre_final, x="Ïû•Î•¥", y="ÏÑ†Ìò∏ÎèÑ_weighted", hue="Ïû•Î•¥", dodge=False)
#show_values(plt1)
plt.title("Weighted Preference by Genre", size=18)

#ÏïÑÏâ¨Ïö¥ Ï†êÏùÄ ascending order Î°ú Ï†ïÎ†¨ÌïòÎäî Îç∞ Ïã§Ìå®ÌñàÎã§Îäî Ï†ê
```
