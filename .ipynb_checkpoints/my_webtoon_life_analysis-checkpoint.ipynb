# 📌2. 완결작 중 내가 완독한 작품 수와 공통된 특징들은?

```python
df.head()
```

```python
#분석 전 이상치 확인
#휴재중인데 완결도 냈다는 건 불가능하니 혹시 잘못 입력된 값이 있는지 확인한다 
df[(df["완결여부"] == 1) & (df["휴재여부"] == 1)]
```

```python
#확인해보니 완결이 아니라 장기 휴재이기 때문에 완결이 아니므로 완결여부를 true(1)에서 false(0)로 변경한다

#row 인덱스 이용하여 "완결여부" 컬럼값 변경
index = df[(df["완결여부"] == 1) & (df["휴재여부"] == 1)].index
df.loc[index,["완결여부"]] = 0

df[(df["완결여부"] == 1) & (df["휴재여부"] == 1)] #변경 성공
```

```python
df_completed = df[(df["완결여부"]==1) & (df["최근 회차"] == df["총 연재 회차"])].copy()
df_completed
```

```python
#완결작 중 내가 완독한 작품수는 13개이다 

len(df_completed["웹툰명"]) 
```

### ※ 인사이트: 완결작 중 내가 완독한 작품수는 13개이다.

```python
#완독여부를 나타내는 파생변수 생성 (true:1 false:0)

#웹툰명이 df_completed에 리스트되어 있다면,
#새 파생변수 "완독여부"의 값을 1로 설정한다
df.loc[df["웹툰명"].isin(df_completed["웹툰명"]),["완독여부"]] = 1

#1이 아닌 나머지값은 null(np.nan)값일 것이므로, 0으로 바꿔준다
df["완독여부"] = df["완독여부"].fillna(0)
df.head()
```

##  2-1. 총 회차 중 내가 읽은 회차의 비율로 웹툰 '선호도'를 측정할 수 있다고 가정한다면, 특징들의 관계는 어떻게 나타날까? 

```python
#파생변수: 선호도
#총 회차 중 내가 읽은 회차의 비율

df["선호도"] = round(df["최근 회차"]/df["총 연재 회차"], 2)
df.head()
```

### 2-1a. 마지막 조회일이 오래됐을 수록 선호도가 낮을까? 
* 별로 취향이 아니였기 때문에 조회한지 오래된 게 아닐까?

```python
#그래프 왼쪽에 나타내기 위해 음수값으로 변환
df["조회일"] = -abs(df["조회일"])
df["조회일"]
```

```python
df_outdated = df[["웹툰명", "조회일", "선호도", "장르", "완결여부"]]
df_outdated
```

```python
sns.scatterplot(data=df_outdated, x="조회일", y="선호도")
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### ※ 인사이트: 위 그래프 상으로는, No correlation. 조회일과 선호도는 상관관계가 없어보인다. 
### 완결이 오래전에 됐다면 선호도가 높아도 조회일이 오래됐을 것이니, 완결된 작품은 제외하고 비교해야 한다.

```python
#완결여부가 true인 것 제거, false 인 것만 
df_outdated1 = df_outdated[df_outdated["완결여부"]==0]
df_outdated1
```

```python
sns.scatterplot(data=df_outdated1, x="조회일", y="선호도", hue="장르")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### ※ 인사이트: 여전히 조회일과 선호도는 상관관계가 크지 않아보인다.
### 결론: 조회일이 오래됐다고 해서 선호도가 낮은 것은 아니다.

### 2-1b. 웹툰의 주요 특징 (장르, 요일) 별 선호도는 어떻게 다를까? 

### [ 장르 ] 

```python
df_genre = round(df.groupby("장르")["선호도"].mean(),3).reset_index().sort_values("선호도")
```

```python
plt.figure(figsize=(16,8))
plt1 = sns.barplot(data=df_genre, x="장르", y="선호도", hue="장르", dodge=False, errwidth=0)
#show_values(plt1)
plt.title("Average Preference by Genre in Ascending Order", size=18)
```

### ※ 인사이트:
### 이 그래프에 따르면, 내가 가장 선호하는, 웹툰이 완결될 때까지 중간에 하차하지 않는 웹툰 장르는 판타지 이다. 스포츠와 개그 웹툰 또한 변심하지 않고 보는 편이다.
### 내가 보는 웹툰 중 비중이 제일 높은 드라마와 로맨스가 앞선 세 장르보다는 '완주'비율이 낮은 편이다. 더 많이 보기 때문에 봐야할 웹툰이 많아서 전개나 스타일이 나랑 안 맞는 경우도 많았을 거라 생각한다.
### 또한, 나는 웹툰을 머리를 비우거나 기분 전환하고 싶을 때 주로 찾기 때문에, 좀 더 감정을 몰입해야하는 드라마, 로맨스, 그리고 스릴러 장르와 같은 것보다 나는 감정 소모를 덜 하고 가볍게 볼 수 있는 개그, 스포츠, 판타지 장르를 더 오래 보는 편인 것 같다. 

## 그러나, 여기서 생각해야할 것은 내가 보는 웹툰 중 해당 장르가 차지하는 비율이다. 내가 33개의 드라마 장르 웹툰을 보고 1개의 판타지 장르 웹툰을 보는데 그 웹툰 하나를 완독했다고 해서 판타지 웹툰 전체를 33개의 웹툰을 보고 있는 드라마 장르보다 더 선호한다고 일반화 하긴 이르다. 

### 2-1c. 파생변수를 가중치 삼아 또 하나의 파생변수를 만든다면?

```python
#장르 빈도수 / 장르 총합 = 장르 비율
#처음에 계산했던 .value_counts()값으로 비율을 계산한다 
df_genre_counts = pd.DataFrame(round(genre_counts/genre_counts.sum(),3))
df_genre_counts = df_genre_counts.rename(columns={"장르":"장르 비율"})

#df_genre_counts = df_genre_counts.reset_index()
df_genre_counts = df_genre_counts.reset_index().rename(columns={"index":"장르"})
df_genre_counts
```

```python
df_genre
```

```python
#장르로 합친 후 계산하여 파생변수 생성 
df_genre_final = pd.merge(df_genre_counts, df_genre, how='inner', on='장르')
df_genre_final["선호도_weighted"] = df_genre_final["장르 비율"] * df_genre_final["선호도"]
df_genre_final
```

```python
plt.figure(figsize=(16,8))
plt1 = sns.barplot(data=df_genre_final, x="장르", y="선호도_weighted", hue="장르", dodge=False)
#show_values(plt1)
plt.title("Weighted Preference by Genre", size=18)

#아쉬운 점은 ascending order 로 정렬하는 데 실패했다는 점
```

### ※ 인사이트: 
### 아까와 완전히 다른 그래프를 그린다. 완독 비율(선호도)이 가장 높았던 판타지가 9번째 선호되는 장르고, 스포츠, 개그에 밀렸던 로맨스와 드라마가 압도적으로 선호됨을 알 수 있다.
### 읽는 웹툰 중 35%씩 차지하는 압도적인 양이 영향을 미친 것으로 보인다.
### 또 한 가지 흥미로운 점은 동등하게 33개씩의 비중을 차지했던 로맨스와 드라마가 가중치가 가미된 선호도에서는 차이를 보였다는 점인데, 나는 드라마 장르의 웹툰보다도 로맨스 장르의 웹툰을 좋아하나보다. 

```python
#pivot_table 사용해서 grouped barplot으로 시각화
df_final = pd.pivot_table(df_genre_final,
                          values=["선호도", "선호도_weighted"],
                          index=["장르"]).sort_values(by="선호도_weighted")
df_final
```

```python
plt.rcParams["figure.figsize"] = (10, 5)
df_final.plot.barh(width=1)
plt.title("장르별 선호도(선호도_weighted)")
```

### ※ 인사이트: 전 후 차이를 보니, 보다 정밀한 분석을 위해서 상황에 따른 가중치를 부여하는 것은 매우 중요하다는 것을 알 수 있다.

# 결론

## 정확한 분석을 위해 상황에 맞는 가중치를 부여해야한다. 
    * 장르별, 요일별 나의 선호도를 더 정밀하게 분석하기 위해 각각 다른 성질의 가중치를 곱했다.
    * 장르별에는 전체 수 대비 각 장르의 비율,
    * 요일별에는 요일별 웹툰 갯수를 곱했다. 
    

## 웹툰에 대한 feature 값을 더 다양하게 탐색해볼 필요가 있다
    *'평균 별점' - 독자의 선호도에 대한 정답 데이터가 될 수 있음, 상관계수 분석에 용이
    *'그림체' - 가설 검증 : 독자의 선호도가 그림체 종류의 영향을 받는다 
    *'인기순위' - 가설 검증 : 군중심리에 따라'인기순'에 노출된 순위가 개인 선호도에 영향을 미친다
    
## 한 사람(나)에 대한 데이터를 수집한 것이기 때문에 규모가 작고 한정적이며, 일반화가 절대 불가능하다. 
    * 그러나 머신러닝 모델을 활용해 나의 개인의 선호도를 예측해볼 수 있다
    
    ---> ex. '신작' 카테고리의 웹툰에 대한 데이터셋을 구축하고, train/ cross validation/ test를 통해
    지도학습을 진행하여 정확도 있는 모델을 확보한다면, 신작 중 어떤 웹툰들을 선호할 것인지 예측해볼 수 있다! 
    
    * 다른 사용자 전체의 데이터가 있으면 더 다양한 분석과 비교가 가능할 것이다
    
## 총 회차 중 내가 읽은 회차의 비율로 웹툰 '선호도'를 판단하는 것은 신뢰도가 부족하다. 
    * 전체 회차를 다 읽지 못하고 중단할 때에는 여러 변수가 작용할 수 있다.
    * 그리고 끝까지 다 완독한다고 해서 제일 좋아하는 웹툰이라는 것도 아니다, 그저 뒷이야기가 궁금해서 완주할 수도 있다. 선호도를 결정짓는 기준이 어렵고 애매한 것 같다. 
    
## 그렇다면 웹툰의 "선호도", 특히 네이버 웹툰 앱에 AI추천 기반이라는 선호도 %는 어떤 feature 들과 변수들로, 어떤 기준으로 결정되는 것일까?
